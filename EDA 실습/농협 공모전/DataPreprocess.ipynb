```python
import re

# 함수: 데이터프레임의 문자열에서 불법 문자를 제거하는 함수
def remove_illegal_characters(df):
    # 데이터프레임의 모든 셀을 순회하며 문자열에 대한 처리
    for col in df.columns:
        if df[col].dtype == 'object':  # 문자열 데이터에 대해서만 처리
            df[col] = df[col].apply(lambda x: re.sub(r'[^\x20-\x7E]+', '', x) if isinstance(x, str) else x)
    return df
```

```python
# preprocess_data 폴더 경로 설정
preprocess_data_path = os.path.join(three_levels_up, "preprocess_data")

# preprocess_data 폴더가 없으면 생성
if not os.path.exists(preprocess_data_path):
    os.makedirs(preprocess_data_path)

# data_name 리스트 안의 데이터프레임을 csv와 excel로 저장
for df_name in data_name:
    # df_name을 변수로 변경하여 데이터프레임 가져오기
    df = globals()[df_name]
    
    #결과 임시 출력
    print(df_name)
    print(df.head())
    
    # 파일 저장 경로 설정
    csv_path = os.path.join(preprocess_data_path, f"{df_name}.csv")
    excel_path = os.path.join(preprocess_data_path, f"{df_name}.xlsx")
    
    # CSV 파일로 저장
    df.to_csv(csv_path, index=False)
    
    # 데이터프레임에서 불법 문자를 제거한 후 Excel 파일로 저장
    df_cleaned = remove_illegal_characters(df)
    df_cleaned.to_excel(excel_path, index=False)


    print(f"{df_name} 저장 완료: CSV({csv_path}), Excel({excel_path})")
```
